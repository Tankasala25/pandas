```python
import pandas as pd
```

```python
People={
    'first':['Rao','ajay','vijay'],
    'last':['Tankasala','Johnson','Johnson'],
    'email':['raotankasala@gmail.com','ajayjohnson@gmail.com','vijaydart@gmail.com']
}
```

```python
df=pd.DataFrame(People)
```

```python
df
```

```python
#this is comparison checking for last name and assigning to variable name
filt= (df['last']=='Johnson')
```

```python
df.loc[filt]
```

```python
# we can also get emails from filtered rows
df.loc[filt,'email']
```

```python
# we can alos use python & , | operators to filter
filt= (df['last']=='Johnson') & (df['first']=='ajay')
```

```python
df.loc[filt]
```

```python
# we can also get opposite of filter bu using '~'
df.loc[~filt]
```
