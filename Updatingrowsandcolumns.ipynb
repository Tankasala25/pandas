```python
'''
In EdLevel, use .replace() to change
"Bachelors degree (BA, BS, B.Eng., etc.)"  "Bachelor"
"Masters degree (M.A., M.S., M.Eng., MBA, etc.)"  "Master".
Then print the first 10 rows of that column.
'''

data['EdLevel'].replace({
'Bachelors degree (BA, BS, B.Eng., etc.)':'Bachelor',
'Masters degree (MA, MS, M.Eng., MBA, etc.)':'Master'},inplace=True)
data['EdLevel']

```

```python
#Filter people from India and in EdLevel "Master"  replace "Master" with "Postgraduate".

filt = (data['Country'] == 'India')
data.loc[filt, 'EdLevel'] = data.loc[filt, 'EdLevel'].replace({'Master': 'Postgraduate'})
```
