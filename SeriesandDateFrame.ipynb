```python
#how to import pandas library
import pandas as pd
```

```python
#let's look  at python dictionary

Person ={
    'first':'Rao',
    'last':'Tankasala',
    'email':'raotankasala@gmail.com'
}
```

```python
#suppose you want same columns for more number of people you can just conver to list.

People={
    'first':['Rao'],
    'last':['Tankasala'],
    'email':['raotankasala@gmail.com']
}
```

```python
#print dictionary
People
```

```python
#you want to get only firstnames from dictionary
People['first']
```

```python
People={
    'first':['Rao','ajay','vijay'],
    'last':['Tankasala','johnson','dart'],
    'email':['raotankasala@gmail.com','ajayjohnson@gmail.com','vijaydart@gmail.com']
}
```

```python
# convert this dictionary as dataframe

df=pd.DataFrame(People)
```

```python
#print df
df
```

In the above cell, we printed the entire DataFrame.  
When you select only **one column** from a DataFrame, it is called a **Series**.  

A **Series** represents a single column along with its rows.  
When you have **two or more columns**, together they form a **DataFrame**.


```python
#let's print one column(series) from dataframe
df['first']
```

```python
#here is the another way to get the column out of dataframe
'''but i prefered first method because with this one there is an ambiguity(confusion) becuase dataframe has method like count,suppose if your table has count column that -
 willcreate a problem'''

df.email
```

```python
#this is dataframe again because it is not printing single column
df[['first','last']]
```

```python
#get the shape of df
df.shape
```

```python
#get columns of dataframe
df.columns
```

```python
#to get rows there are two methods 'loc' and 'iloc'

df.iloc[0]
```

```python
#get two cloumns
df.iloc[[0,1]]
```

```python
#get two rows with one column
df.iloc[[0,1],2]
```

```python
#same with loc too. but iloc only work with indexes(numbers) and loc work with labels(like, first , last ,email)

df.loc[0,'first']

```

```python
df.loc[[0,1],'last']
```

```python
#you can usinng slicing in dataframes to get from particular row to column
#but not like python here slicing will wwork differently
#in python the last number at index is not included but when you are using in dataframe the last one also included.

df.loc[ 0:2 , 'first':'email']
```
