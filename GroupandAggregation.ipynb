```python
data = {
    'Department': ['IT', 'IT', 'HR', 'HR', 'Finance', 'Finance', 'IT', 'Finance', 'HR'],
    'Employee': ['Rao', 'Ajay', 'Vijay', 'Kiran', 'Maya', 'Arun', 'Sita', 'Ravi', 'Sara'],
    'Salary': [80000, 75000, 60000, 58000, 90000, 87000, 95000, 88000, 62000],
    'Experience': [5, 3, 6, 4, 10, 8, 12, 9, 7],
    'Gender': ['M', 'M', 'M', 'M', 'F', 'M', 'F', 'M', 'F']
}

df = pd.DataFrame(data)
print(df)
```

```python
df.groupby('Department')['Salary'].mean()
```

```python
df.groupby('Department')['Salary']
```

```python
df.groupby('Department')['Salary'].sum()
```

```python
#split-apply function - combine - this is grouping
#Which department has the highest average experience?
```

```python
df.groupby('Department')['Experience'].mean()
```

```python
#How many male vs female employees are in each department?
```

```python
df.groupby('Department')['Gender'].value_counts()
```

```python
#Find top 2 salaries in each department
```

```python
df.groupby('Department')['Salary'].nlargest(2)
```

```python
#Find the average salary of employees with more than 5 years of experience per department
```

```python
df.query("Experience > 5").groupby('Department')['Salary'].mean().round(2)
```

```python
#Show the total salary paid by each department.
```

```python
df.groupby('Department')['Salary'].sum()
```

```python
df.groupby(['Department','Gender'])['Experience'].mean()
```

```python
#Top 2 salaries per department (with employee names) next
df.groupby(['Department','Employee'])['Salary'].nlargest(2)
```

```python
df.columns
```

```python
#“For each department, show total salary, number of employees, highest salary, and lowest salary”?
```

```python
df.groupby('Department')['Salary'].agg(['sum','count','max','min'])
```

```python
#Show the average salary per department using a pivot table.
```

```python
pd.pivot_table(df,index='Department',values='Salary',aggfunc='mean')
```

```python
#Show the average salary by Department and Gender.
```

```python
pd.pivot_table(df,index='Department',columns='Gender',values='Salary',aggfunc='mean')
```

```python
#Display a pivot table that shows the average salary and experience per department.
```

```python
pd.pivot_table(df,index='Department',values=['Salary','Experience'],aggfunc='mean')
```

```python
#Show the average salary of employees with more than 5 years of experience, grouped by Department.
```

```python
pd.pivot_table(df.query("Experience > 5"),index='Department',values='Salary',aggfunc='mean')
```

```python
'''
For each department, show both:

total salary

number of employees
and also show a Grand Total row at the bottom.
'''
```

```python
pd.pivot_table(df,index='Department',values='Salary',aggfunc=['sum','count'],margins=True)
```
