```python
import pandas as pd
```

```python
People={
    'first':['Rao','ajay','vijay'],
    'last':['Tankasala','johnson','dart'],
    'email':['raotankasala@gmail.com','ajayjohnson@gmail.com','vijaydart@gmail.com']
}
```

```python
df=pd.DataFrame(People)
```

```python
df
```

```python
#fetch the email column
df['email']
```

```python
#An **index** in pandas uniquely identifies each row — like a “key” in a database.
#suppose you want to make email column as index

df.set_index('email')
```

```python
#now we print only dataframe
df
```

```python
#even though we set index for email column it is not reflected in original data frame . to excute that we use inplace=True
df.set_index('email',inplace=True)
```

```python
#now we if print dataframe you can see the changes.
```

```python
# we can also check like this in dataframe which was indexed?

df.index
```

```python
# we can also fetch row using email as index
df.loc['raotankasala@gmail.com']
```

```python
#you want last name using email as index

df.loc['raotankasala@gmail.com','last']
```

```python
# this will get you key error becuase the loc is looking for index '0' we have no more numbers as index becuase we set email as index
df.loc[0]
```

```python
#this will fetch data even though we gave number it will fetch data
df.iloc[0]
```

```python
#to reset index
df.reset_index(inplace=True)
```

```python
#now if print dataframe it will come as original
df
```
