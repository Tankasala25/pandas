```python
import pandas as pd
```

```python
People={
    'first':['Rao','ajay','vijay'],
    'last':['Tankasala','Johnson','Johnson'],
    'email':['raotankasala@gmail.com','ajayjohnson@gmail.com','vijaydart@gmail.com']
}
```

```python
df=pd.DataFrame(People)
```

```python
df['first'] + ' '+df['last']
```

```python
#to add two column
df['full_name'] = df['first'] + ' '+df['last']
```

```python
#to drop columns
df.drop(columns=['first','last'], inplace=True)
```

```python
#split the full_name as first and last
df['full_name'].str.split(' ',expand=True)
```
