```python
dat = {
    'Name': ['Rao', 'Ajay', 'Vijay', 'Kiran', 'Divya'],
    'Country': ['India', 'USA', 'India', 'Canada', 'India'],
    'YearsCode': [12, 5, 3, 8, 15],
    'Language': ['Python', 'Java', 'C++', 'Python', 'JavaScript']
}

df = pd.DataFrame(dat, index=[101, 102, 103, 104, 105])
print(df)
```

```python
df
```

```python
#Change Ajay’s Country (index 102) to "United States".
```

```python
df['Country'].replace({'USA':'United States'},inplace=True)
```

```python
df
```

```python
#“Change Vijay’s YearsCode (index 103) to 10”?
```

```python
df['YearsCode'].replace({3:10},inplace=True)
```

```python
df
```

```python
#"Change Kiran’s entire row (index 104) to ['Kiran', 'Canada', 9, 'C#']”
```

```python
df.loc[104]=['Kiran','Canada',9,'C#']
```

```python
df
```

```python
#For all rows where Country == 'India', set Language = 'Python'”
```

```python
filt=(df['Country']=='India')
df.loc[filt,'Language']='Python'
```

```python
df
```

```python
#Add a new column ExperienceLevel — 'Senior' if YearsCode >= 10, else 'Junior' (in one line)”
```

```python
df['ExperienceLevel']=df['YearsCode'].apply(lambda x:'Senior' if x>=10 else 'Junior')
```

```python
df
```

```python
#For all Senior developers in India, update Country → 'India (Experienced)' and Language → 'Python (Advanced)'”
```

```python
filt=(df['ExperienceLevel']=='Senior')
df.loc[filt,['Country','Language']]=['India (Experienced)','Python (Advanced)']
```
